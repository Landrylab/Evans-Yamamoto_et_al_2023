```r
library(dplyr)
library(tidyr)
library(tidyverse)

library(ggplot2)
library(ggtree)
library(castor)
library(treeio)
library(TreeTools)

library(gridExtra)
library(cowplot)
library(egg)
library(grid)

library(devtools)

session_info()
```

```r
# Tree from 
# Li Y, Steenwyk JL, Chang Y, Wang Y, James TY, Stajich JE, Spatafora JW, Groenewald M, Dunn CW, Hittinger CT, et al. 2021. A genome-scale phylogeny of the kingdom Fungi. Curr. Biol. 31:1653–1665.e5.
tree_f = "./tree/1672taxa_290genes_bb_1.treefile"
```

```r
tree = read.newick(tree_f)
```

```r
names(tree)
```

```r
# List species in screening

sp_list = c("Saccharomyces_pastorianus",
            "Saccharomyces_cerevisiae",
            "Saccharomyces_eubayanus",
            "Saccharomyces_bayanus",
            "Saccharomyces_uvarum",
            "Saccharomyces_paradoxus",
            "Saccharomyces_mikatae",
            "Saccharomyces_kudriavzevii",
            "Candida_glabrata", # identical to "Nakaseomyces_glabratus"
            "Kazachstania_saulgeensis",
            "Kazachstania_barnettii",
            "Kazachstania_exigua",
            "Kazachstania_africana",
            "Kazachstania_naganishii",
            "Naumovozyma_castellii",
            "Naumovozyma_dairenensis",
            "Vanderwaltozyma_polyspora",
            "Tetrapisispora_phaffii",
            "Tetrapisispora_blattae",
            "Torulaspora_delbrueckii",
            "Torulaspora_globosa",
            "Zygotorulaspora_mrakii",
            "Zygosaccharomyces_parabailii",
            "Zygosaccharomyces_bailii",
            "Zygosaccharomyces_rouxii",
            "Lachancea_meyersii",
            "Lachancea_lanzarotensis",
            "Lachancea_dasiensis",
            "Lachancea_nothofagi",
            "Lachancea_quebecensis",
            "Lachancea_thermotolerans",
            "Lachancea_fermentati",
            "Lachancea_mirantina",
            "Kluyveromyces_marxianus",
            "Kluyveromyces_lactis",
            "Eremothecium_gossypii",
            "Eremothecium_cymbalariae"
            )


```

```r
list_in_tree = tree$tip.label[ tree$tip.label %in% sp_list]
length(sp_list)
length(list_in_tree)

#sp_list[!(sp_list %in% c(list_in_tree))]
list_in_tree
```

```r
tree_node2tiplab = data.frame(node=1:length(tree$tip.label),label = tree$tip.label)

tree_node2tiplab= tree_node2tiplab %>% rowwise() %>% mutate(selected = if_else( label %in% list_in_tree,1,0))

sel_nodes_df =tree_node2tiplab%>%filter(selected==1)
sel_nodes    = sel_nodes_df$label

head(tree_node2tiplab)
length(sel_nodes)
```

```r
sel_nodes
```

```r
tree_sel=  tree_subset(
                      tree=tree,
                      node=sel_nodes,
                      levels_back=6,
                      root_edge = FALSE
                        )
tree_sel
```

```r
tree_sel$tip.label


species = data.frame(node=1:length(tree_sel$tip.label),label = tree_sel$tip.label)

write.csv(species,"./tree/sel_species.csv",row.names=FALSE)
```

```r
bs = data.frame(nodename=1:tree_sel$Nnode,bootstrap= tree_sel$node.label, stringsAsFactors = F)

rownames(bs) <- NULL
head(bs)

nodesupport = c(rep(0,length(tree_sel$tip.label)),  as.numeric(tree_sel$node.label))
```

```r
tree_p= ggtree(tree_sel)
tree_p
```

```r
tree_p= ggtree(tree_sel)
tree_p <- tree_p %<+% bs + geom_tiplab(align = TRUE,linesize=0)
tree_p <- tree_p + geom_text(aes(label = bootstrap), hjust = -0.4, vjust = -0.4, size = 3) + geom_nodelab(aes(label = nodesupport))+
        #theme_tree2(legend.position=c(.1,.9))+
        geom_treescale(x=0, linesize = 1,fontsize = 5,family = "Helvetica",color="#000000")+
        ggplot2::xlim(0, 1.5)+
        theme(legend.position="left",legend.title=element_text(color="#000000"))+
        labs(color = "Fast Bootstrap")+
        coord_cartesian(clip="off")

tree_p
```

```r
p <-set_panel_size(tree_p,width  = unit(25, "cm"),height = unit(25, "cm")) 
ggsave(plot=p,"./tree/SelectedSpeciesTree_plot.pdf" ,width=unit(20,"cm"),height=unit(20,"cm"))
```

```r
library(dplyr)
library(tidyr)

library(ggplot2)
library(ggtree)
library(castor)
library(treeio)

library(gridExtra)
library(cowplot)
library(egg)
library(grid)
#install.packages("TreeTools")
library("TreeTools")


```

```r
tree
UnrootTree(tree)
```

```r
tree_f = "./alignments_mafft/raxml-ng/HRR25_mafft_translatorx.nt_ali.fasta.raxml.besttree"
tree = read.newick(tree_f)
tree
tree_p=ggtree(tree) + theme_tree2(legend.position=c(.1,.9))+ geom_text(aes(label=node), hjust=-.2)+
geom_treescale(x=0, linesize = 1,fontsize = 5,family = "Helvetica",color="#000000")+
geom_tiplab(align = TRUE,linesize=0)+ggplot2::xlim(0,12)+
theme(legend.position="left",legend.title=element_text(color="#000000"))
tree_p
p <-set_panel_size(tree_p,width  = unit(50, "cm"),height = unit(12, "cm")) 
ggsave(plot=p,"./alignments_mafft/raxml-ng/HRR25_mafft_translatorx.nt_ali.fasta.raxml.besttree.pdf",limitsize = FALSE ,width=unit(20,"cm"),height=unit(10,"cm"))

```

```r
tree_f_nogap = "./alignments_mafft_nogaps/raxml-ng/HRR25_mafft_translatorx_nt_ali_nogaps.fasta.raxml.bestTree"
tree_nogap = read.newick(tree_f_nogap)
tree_nogap
tree_p_nogap=ggtree(tree_nogap) + theme_tree2(legend.position=c(.1,.9))+ geom_text(aes(label=node), hjust=-.2)+
geom_treescale(x=0, linesize = 1,fontsize = 5,family = "Helvetica",color="#000000")+
geom_tiplab(align = TRUE,linesize=0)+ggplot2::xlim(0,12)+
theme(legend.position="left",legend.title=element_text(color="#000000"))
tree_p_nogap
p <-set_panel_size(tree_p_nogap,width  = unit(50, "cm"),height = unit(12, "cm")) 
ggsave(plot=p,"./alignments_mafft_nogaps/raxml-ng/HRR25_mafft_translatorx_nt_ali_nogaps.fasta.raxml.bestTree.pdf",limitsize = FALSE ,width=unit(20,"cm"),height=unit(10,"cm"))
tree_nogap$edge.length = NULL
write.tree(tree_nogap,"../01_protocol_CODEML/nogap/HRR25_mafft_translatorx_nt_ali_nogaps.fasta.raxml.bestTree_nobl.txt")
```
